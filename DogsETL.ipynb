Importando bibliotecas

```python
import pandas as pd
import numpy as np
import chardet
```

Detectando o encoding

```python
import chardet

# Fun√ß√£o para detectar a codifica√ß√£o de um arquivo
def detectar_codificacao(arquivo):
    with open(arquivo, 'rb') as f:
        data = f.read()
    encoding_result = chardet.detect(data)
    encoding = encoding_result['encoding']
    return encoding

# Detectando codifica√ß√£o para 'dog_breeds.csv'
encoding_breeds = detectar_codificacao('dog_breeds.csv')
print("Detected Encoding for dog_breeds.csv:", encoding_breeds)
```

Observando as primeiras 5 linhas da tabela


```python
dados = pd.read_csv('dog_breeds.csv',encoding='ascii', sep=',') #separador
dados.head(5)
```

Visualizando o tamanho do dataset

```python
dados.shape #visualizar numero de linhas/registros e colunas/variaveis
```

Observando se tem valores duplicados


```python
dados.Breed.nunique()
dados.Breed.value_counts()
```

Retirando as duplicatas

```python
dados.drop_duplicates(subset=['Breed'], inplace=True)
dados.shape
```

Fazendo uma c√≥pia do dataset para realizar manipula√ß√µes

```python
dados_copia = dados.copy()
dados_copia.shape

dados_copia.head(10)
```

Mudando o nome de algumas colunas

```python
# Remover espa√ßos extras dos nomes das colunas
dados_copia.columns = dados_copia.columns.str.strip()

# Renomear as colunas
dados_copia = dados_copia.rename(columns={'Country of Origin': 'Origin','Common Health Problems': 'Common Disease'})
```

```python
dados_copia.head(10)
```

Separando a coluna Height (in) em Height_min e Height_max

```python
dados_copia[['Height_min', 'Height_max']] = dados_copia['Height (in)'].str.split('-', expand=True)
dados_copia.head(5)
```

Deletando a coluna Height (in)

```python
dados_copia = dados_copia.drop(columns=['Height (in)'])
dados_copia.head(5)
```

Mudando o tipo de altura para float


```python
dados_copia['Height_min'] = pd.to_numeric(dados_copia['Height_min'], downcast='float')
dados_copia['Height_max'] = pd.to_numeric(dados_copia['Height_max'], downcast='float')
```

Transformando o tipo de Height_min e Height_max com medi√ß√µes de polegar para cent√≠metros(* 2.54)


```python
dados_copia['Height_min'] = round(dados_copia['Height_min']*2.54, 2 ) # N√£o consegui arredondar os valores
dados_copia['Height_max'] = round(dados_copia['Height_max'] *2.54,2 )
dados_copia.head(5)
```

Separando a coluna 'Longevity (yrs) em 'Longevity_min e 'Longevity_max


```python
dados_copia[['Longevity_min', 'Longevity_max']] = dados_copia['Longevity (yrs)'].str.split('-', expand=True)
dados_copia.head(5)
```

Deletando a coluna Longevity (yrs)

```python
dados_copia= dados_copia.drop(columns=['Longevity (yrs)'])
```

Transformando o tipo de Longevity_min e Longevity_max em num√©rico

```python
dados_copia['Longevity_min']= pd.to_numeric(dados_copia['Longevity_min'])
dados_copia['Longevity_max']= pd.to_numeric(dados_copia['Longevity_max'])
dados_copia.head(10)
```

Padronizando os valores dos campos espec√≠ficos para min√∫sculo


```python
dados_copia['Fur Color'] = dados_copia['Fur Color'].str.lower()
dados_copia['Color of Eyes'] = dados_copia['Color of Eyes'].str.lower()
dados_copia['Character Traits'] = dados_copia['Character Traits'].str.lower()
dados_copia['Common Disease'] = dados_copia['Common Disease'].str.lower()

dados_copia.head(10)
```

Revisando os tipos

```python
dados_copia.dtypes
```

Separando a coluna de Common Disease e adicionando os valores bin√°rios para presen√ßa e aus√™ncia


```python
disease_dummies = dados_copia['Common Disease'].str.get_dummies(sep=', ')
dados_copia = pd.concat([dados_copia, disease_dummies], axis=1)
```

```python
dados_copia.head(5)
```

Separando a coluna Character Traits e adicionando os valores bin√°rios para presen√ßa e aus√™ncia

```python
traits_dummies = dados_copia['Character Traits'].str.get_dummies(sep=', ')
dados_copia = pd.concat([dados_copia, traits_dummies], axis=1)
```

```python
dados_copia.head(5)
```

Apagando Colunas que n√£o ser√£o utilizadas nas an√°lises

```python
dados_copia= dados_copia.drop(columns=['Character Traits', 'Common Disease', 'Fur Color', 'Color of Eyes'])
```

```python
dados_copia.head(5)
```
